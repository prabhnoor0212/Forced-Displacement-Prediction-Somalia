```python
d1[' Number of Individuals '] = d1[' Number of Individuals '].apply(lambda x: int(x.replace(',','')))
```

```python
def people_arriving(lat, lon, date):
    return d1[(d1['lat']==lat) & (d1['lon']==lon) & (d1['date']==date)][' Number of Individuals '].sum()
```

```python
people_arriving(d2.iloc[0]['lat'], d2.iloc[0]['lon'],d2.iloc[0]['date'])
```

```python
d1[(d1['lat']==d2.iloc[100]['lat']) & (d1['lon']==d2.iloc[100]['lon']) & (d1['date']==d2.iloc[100]['date'])][' Number of Individuals '].sum()
```

```python
from tqdm import tqdm
mothly_arrivals = []
no_arrival_regions = set()
for _, row in tqdm(d2.iterrows()):
    arrivals = people_arriving(row['lat'], row['lon'], row['date'])
    if arrivals==0:
         no_arrival_regions.add(row['place'])
    mothly_arrivals.append(arrivals)
```

```python
d2['monthly_arrivals'] = mothly_arrivals
```
